```python
import pickle
import networkx as nx
from pprint import pprint as myprint
import matplotlib.pylab as plt
plt.rcParams["font.sans-serif"]=["SimHei"]
```

```python
trigram_b = open(r"../tmp/trigram.dat","rb")
triple_set = pickle.load(trigram_b)
myprint(triple_set)
```

第一个很显眼，居然是空格，说明分词的部分可能有问题，必要的时候返回解决

```python
node_set = set()
rel_set = set()
for sbj,verb,obj in triple_set:
    node_set.add(sbj)
    node_set.add(obj)
    rel_set.add((sbj,obj))
myprint(len(rel_set))
myprint(len(node_set))
```

```python
graph = nx.Graph()
graph.add_nodes_from(node_set)
graph.add_edges_from(rel_set)
```

```python
nx.draw(graph,with_labels=True)
plt.show()
```
