```python
import re
from pprint import pprint as myprint
from bs4 import BeautifulSoup as bs
```

```python
file = open(r"../resource/信息资源建设讲义.txt","r")
raw_text_list = [line.strip() for line in file.readlines() if line.strip() != ""]
myprint(raw_text_list[:20])
file.close()
```

打开资源文件并将其以换行为分隔符读入字符串列表

```python
index = raw_text_list.index("4、肖希明信息资源建设的变革与发展国家图书馆出版社2010")
raw_text_list =  raw_text_list[index+1:]
```

去除资源文件中的课程信息以及大纲部分，这一部分有其利用价值，但是目前不考虑

```python
text_list = []
[text_list.append(x) for x in raw_text_list if x not in text_list]
myprint(text_list[:20])
```

去除重复行，在课件中会出现同一个标题为了显示美观而在多个页面上出现的情况，因此需要处理。但是也有可能将本应重复的部分删除。

```python
pstring1 = r"^第[一二三四五六七八九十]+章"
pstring2 = r"^第[一二三四五六七八九十]+节"
pstring3 = r"^[一二三四五六七八九十]+[、]"
pstring4 = r"^[1234567890]+[、]"
pstring = [pstring1,pstring2,pstring3,pstring4]
pattern1 = re.compile(pstring1)
pattern2 = re.compile(pstring2)
pattern3 = re.compile(pstring3)
pattern4 = re.compile(pstring4)
pattern = [pattern1,pattern2,pattern3,pattern4]

pattern_all = re.compile(r"^第[一二三四五六七八九十]+章|^第[一二三四五六七八九十]+节|^[一二三四五六七八九十]+[、]|^[1234567890]+[、]")
```

设计了正则表达式用来提取各个大纲级别的大纲，这个提取仅仅适用于目前的试验文档，对于不同的文档，仍然有必要写不同的正则表达式。

```python
bigram_list = []
for line in text_list:
    if not pattern_all.match(line):
        tmp = (line,0)
        bigram_list.append(tmp)
    for i in range(len(pattern)):
        if pattern[i].match(line):
            tmp = (re.sub(pstring[i],"",line),i+1)
            bigram_list.append(tmp)
myprint(bigram_list[:20])
```

试验上面所写的正则表达式提取大纲的效果
关于大纲级别：
0 -> 正文
1 -> 一级标题
2 -> 二级标题
3 -> 三级标题
4 -> 四级标题

```python
new_bigram_list = []
for i,tmp in enumerate(bigram_list):
    s,r = tmp
    if r!=0:
        new_bigram_list.append(tmp)
    else:
        pre_s,pre_r = new_bigram_list[-1]
        new_bigram_list[-1] = (pre_s+s,pre_r)
myprint(new_bigram_list[:20])
```

将正文的内容归并到其前一个标题，在逻辑上这是合适的，并且减少了另外处理正文的工作量

```python
bigram_file = open(r"../tmp/bigram.txt","w",encoding="utf-8")
for s,r in new_bigram_list:
    bigram_file.write(s+","+str(r)+"\n")
bigram_file.close()
```
