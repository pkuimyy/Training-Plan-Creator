```python
import codecs
import re
import pprint
myprint = pprint.pprint
from bs4 import BeautifulSoup as bs
```

```python
file = codecs.open(r"../resource/信息资源建设讲义.txt","r")
```

```python
raw_text = file.read()
# print(raw_text)
```

上面的原始文本需要注意一些地方：
1. 课程信息怎么处理
2. 客套话怎么处理
3. 粒度要小到什么程度
4. 序号怎么处理
5. "作业"、"思考题"怎么处理
6. 有重复的文本

目前的处理方法如下：
1. 课程信息去除
2. 客套话并不多，不做处理
3. 处理两级标题
4. 序号同上，处理两级标题，处理细节见正则表达式
5. 保留，认为属于课程内容，将来有需要的时候可以提取出来
6. 重复部分只保留第一处

```python
raw_text_list = [x for x in raw_text.split("\n") if x != ""]
myprint(raw_text_list)
```

```python
index = raw_text_list.index("4、肖希明信息资源建设的变革与发展国家图书馆出版社2010")
raw_text_list = raw_text_list[index+1:]
```

```python
text_list = []
[text_list.append(x) for x in raw_text_list if x not in text_list]
myprint(text_list)
```

除去重复的行，这里是有风险的

```python
pstring1 = r"^第[一二三四五六七八九十]+章"
pstring2 = r"^第[一二三四五六七八九十]+节"
pstring3 = r"^[一二三四五六七八九十]+[、]"
pstring4 = r"^[1234567890]+[、]"
pstring = [pstring1,pstring2,pstring3,pstring4]
pattern1 = re.compile(pstring1)
pattern2 = re.compile(pstring2)
pattern3 = re.compile(pstring3)
pattern4 = re.compile(pstring4)
pattern = [pattern1,pattern2,pattern3,pattern4]

pattern_all = re.compile(r"^第[一二三四五六七八九十]+章|^第[一二三四五六七八九十]+节|^[一二三四五六七八九十]+[、]|^[1234567890]+[、]")
```

提取四级大纲，用正则表达式完成

```python
for line in text_list:
    if pattern_all.match(line):
        print(line)
```

文件中第632行左右，出现了序号与内容不连续的情况。如1.
balabala
另外，原始文本中标号错误的时候。


```python
res_file = open(r"../tmp/信息资源建设.tuple","w")
text_tuple_list = []
for line in text_list:
    if not pattern_all.match(line):
        tmp = (line,0)
        text_tuple_list.append(tmp)
        res_file.write(str(tmp)+"\n")
    for i in range(len(pattern)):
        if pattern[i].match(line):
            tmp = (re.sub(pstring[i],"",line),i+1)
            text_tuple_list.append(tmp)
            res_file.write(str(tmp)+"\n")
myprint(text_tuple_list)
res_file.close()
```

可能笨拙，但是确实有效

```python
new_tuple_list = []
for i,tmp in enumerate(text_tuple_list):
    s,r = tmp
    if r != 0:
        new_tuple_list.append(tmp)
    else:
        pre_s,pre_r = new_tuple_list[-1]
        new_tuple_list[-1] = (pre_s + s,pre_r)
myprint(new_tuple_list)
```

这里将大纲级别为正文的行合并到其指直属的标题，这是建立在某标题下的正文是该标题的解释语句的假设的基础之上的

```python
html_string = ""
for s,r in new_tuple_list:
    html_string += r"<h%d>%s<h%d>"%(r,s,r)
print(html_string)
```

```python
soup = bs(html_string,"html5lib")
print(soup.prettify())
res_file = open(r"../tmp/信息资源建设html5lib.html","w")
res_file.write(soup.prettify())
res_file.close()
```
