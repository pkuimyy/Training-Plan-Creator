```python
import thulac
#http://thulac.thunlp.org/
import codecs
import nltk
```

```python
thu = thulac.thulac(seg_only=True)
```

```python
f = open("信息资源建设讲义.txt","r")
raw_text = f.read()
#print(raw_text)
```

```python
cutted_text = thu.cut(raw_text)
#print(text)
```

```python
stoplist1 = [x.strip() for x in codecs.open("哈工大停用词表.txt","r","utf8").readlines()]
```

```python
stoplist2 = [x.strip() for x in codecs.open("百度停用词表.txt","r","utf8").readlines()]
```

```python
stoplist3 = [x.strip() for x in codecs.open("中文停用词表.txt","r","utf8").readlines()]
```

```python
stoplist4 = [x.strip() for x in codecs.open("四川大学机器智能实验室停用词库.txt","r","utf8").readlines()]
```

```python
stopset = set(stoplist1 + stoplist2 + stoplist3 + stoplist4)
```

```python
print(cutted_text)
```

```python
cutted_text = [x[0] for x in cutted_text]
```

```python
print(cutted_text)
```

```python
# cutted_text = set(cutted_text)
```

```python
text = [x for x in cutted_text if x not in stopset]
```

```python
print(text)
```

```python
fd = nltk.FreqDist(text)
```
